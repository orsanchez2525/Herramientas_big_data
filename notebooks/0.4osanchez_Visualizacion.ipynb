```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#scatter plot (diagramas de dispersión, dos variables)
from turtle import title


x = np.arange(1,20)
y = x**2
y1=x**2 + x*5

plt.scatter(x,y,s=50,color='black',marker='s',label='x ele 2')
plt.scatter(x,y1,s=50,color='red',marker='s',label='x ele 2 + 5x')
plt.legend(title='Funciones')
plt.title('***Tipos de funciones***',fontsize=20,c='Blue')
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.xlim(0,10)
plt.grid()
#plt.text()

#Agrgar coordenadas marcadores en texto 
for val_x in x:
    plt.text(val_x,val_x**2,f'({val_x},{val_x**2})')
```

```python
media=20
str=3
numeros=np.random.normal(media,str,1000)
plt.hist(numeros,color='red',bins=20,cumulative=False,density=False});
plt.title('Distribución normal')
plt.xlabel('Numeros')
plt.ylabel('Frecuencia')
```
